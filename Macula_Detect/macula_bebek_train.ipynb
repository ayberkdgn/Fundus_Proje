```python
import os
import yaml
from ultralytics import YOLO

def main():
    # 1. KRİTİK ADIM: Sıfır model yerine YETİŞKİN modelinin en iyi ağırlıklarını yüklüyoruz!
    # Bu model zaten göz anatomisini çok iyi biliyor.
    yetiskin_model_path = r"C:\Users\Dogan\Desktop\Ayberk Dosyalar\Bitirme Projesi\Fundus\FundusProje\Macula_Detect\runs\detect\yolov8_macula_yetiskin\weights\best.pt"
    model = YOLO(yetiskin_model_path)

    # 2. Sadece BEBEK datalarının olduğu klasör yolu
    bebek_ana_dizin = "C:/Users/Dogan/Desktop/Ayberk Dosyalar/Bitirme Projesi/Fundus/all_fundus_datas/macula_detect_bebek_data"

    # 3. Bebek dataları için geçici bir YAML oluşturuyoruz
    bebek_yaml_icerik = {
        "path": bebek_ana_dizin,
        "train": "train/images",
        "val": "val/images",   # val klasörün yoksa hata vermesin diye train yaptık
        "test": "test/images",  
        "names": {0: "macula"}
    }

    bebek_yaml_yolu = os.path.join(bebek_ana_dizin, "bebek_macula.yaml")
    with open(bebek_yaml_yolu, "w", encoding="utf-8") as f:
        yaml.dump(bebek_yaml_icerik, f, default_flow_style=False)

    print("Bebek için transfer öğrenme eğitimi başlıyor...")
    
    # 4. Eğitimi Başlatma
    results = model.train(
        data=bebek_yaml_yolu,  
        epochs=100,          # Yetişkin modelinden başladığımız için 50-60 epoch fazlasıyla yeterli
        imgsz=640,          
        batch=16,           
        device=0,           
        workers=0,          
        lr0=0.001,          # Öğrenme oranını düşük tutarak yetişkin bilgilerini koruyoruz
        augment=True,       # Albumentations veri artırma (blur, gürültü vb.) alt yapısını açar
        hsv_v=0.4,          # Parlaklık oynamaları (Bebek resimlerindeki flaş patlamalarına karşı)
        mosaic=0.5,           # Bebek resimlerindeki bulanıklığa alışması için blur augmentasyonunu açtık
        name="yolov8_macula_bebek" 
    )
    print("Bebek gözü için özel model başarıyla eğitildi!")

if __name__ == "__main__":
    main()
```
