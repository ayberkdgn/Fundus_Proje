```python
import os
import yaml
from ultralytics import YOLO

def main():
    # 1. Model Seçimi
    model = YOLO("yolov8n.pt")

    # 2. Şu anki bilgisayarına göre mutlak ana dizin yolu
    ana_dizin = "C:/Users/Dogan/Desktop/Ayberk Dosyalar/Bitirme Projesi/Fundus/all_fundus_datas/macula_detect_yetiskin_data"

    # 3. YOLO'nun hata vermemesi için dinamik olarak geçici bir YAML dosyası oluşturuyoruz
    gecici_yaml_icerik = {
        "path": ana_dizin,
        "train": "train/images",
        "val": "val/images",   # val klasörü yoksa hata vermesin diye train yaptık
        "test": "test/images",  # test klasörü yoksa hata vermesin diye train yaptık
        "names": {0: "macula"}
    }

    # YAML dosyasını klasörün içine fiziksel olarak kaydediyoruz
    gecici_yaml_yolu = os.path.join(ana_dizin, "yetiskin_macula.yaml")
    with open(gecici_yaml_yolu, "w", encoding="utf-8") as f:
        yaml.dump(gecici_yaml_icerik, f, default_flow_style=False)
        
    print(f"Yeni ve temiz YAML dosyası oluşturuldu: {gecici_yaml_yolu}")

    # 4. Eğitimi Başlatma (Oluşturduğumuz dosya yolunu veriyoruz)
    print("YOLOv8 Makula Eğitimi Başlıyor...")
    results = model.train(
        data=gecici_yaml_yolu,  # Artık hata alması imkansız bir dosya yolu verdik!
        epochs=100,         
        imgsz=640,          
        batch=16,           
        device=0,           # CUDA için 0 (Ekran kartın desteklemiyorsa 'cpu' yaz)
        workers=0,          # Windows'taki DataLoader kilitlenmelerini önler
        name="yolov8_macula_yetiskin" 
    )
    print("Eğitim başarıyla tamamlandı!")

if __name__ == "__main__":
    main()
```
